```python
P = float(input("Digite el Peso en Kg: "))
T = float(input("Digite la Talla en m: "))
S = 0.2024 * (P**0.425) * (T**0.725)
round(S,2)
```

```python
n = int(input("Digite el nÃºmero de actividades: "))
TT = 0
T = []
C = []
for i in range(n):
    t = float(input(f"{i+1}. Tiempo: "))
    TT += t
    T.append(t)

for i in range(n):
    a = float(input(f"{i+1}. PosiciÃ³n: "))
    b = float(input(f"{i+1}. Tipo de trabajo: "))
    c = (a+b)*T[i]/TT
    C.append(c)

C = sum(C)
GCT = C*60*factor
GCT
```
