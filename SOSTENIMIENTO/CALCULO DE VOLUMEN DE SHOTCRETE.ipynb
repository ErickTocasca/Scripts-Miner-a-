## C√°lculo del per√≠metro

**P = 2H - r(4-Pi) + A**

P = Perimetro(m)

H = Altura(m)

R = Radio(m)

A = Ancho(m)

##  FORMULA 

**V = (PxL) x %Rg x %Rb x %C x Cp x e**

V = Volumen (m3)

P = Perimtetro (m)

L = Longitud (m)

Rg = Rugosidad (%)

Rb = Rebote (%)

C = Cuba (%)

Cp = Compactaci√≥n (%)

e = Espesor (m)

```python
import numpy as np
H = int(input("Altura m: "))
R = int(input("Radio m: "))
A = int(input("Ancho m: "))
L = int(input("Longitud m: "))
Rg = int(input("Rugosidad %: "))
Rb = int(input("Rebote %: "))
C = int(input("Cuba %: "))
Cp = int(input("Compactaci√≥n %: "))
e = int(input("Espesor m: "))

P = 2 * H - R * (4 - np.pi) + A
V = P * L * Rg * Rb * C * Cp * e / 1000000

print(f"El voluemn de ShotCrete lanzado ser√° {V} m3")
```
