Calcular la producción horaria y por 12 turnos de un camión minero CAT 793D, si tuviera las siguientes condiciones:
* Material: suelo arenoso (factor 0.8)
* Densidad de material: 1800 kg/m3
* Capacidad de carga: 229 TM
* Capacidad de volumen: 176 m3
* Trabaja en conjunto con una pala de cables con capacidad de 30 m3 (capacidad colmada)
* Condiciones de operación y mantenimiento: Normales (factor 0.65)

## PASO 1: CALCULAR EL NÚMERO DE CICLOS DE CARGA DEL CAMIÓN
* La carga útil del camión 793D es de 229 Tm de 176 m3 por lo que sí trabaja con una pala de cables de 30 m3 de capacidad colmada y con materiales que dificultan llenar el  cucharon.

**factor de llenado = 0.8**

* Para obtener el número de ciclos, se realiza el cociente de la capacidad de la tolva entre la producción por ciclo de la pala.

```python
q = 30 * 0.8
N = 176 / q
print(f"{round(N)} ciclos")
```

## PASO 2: CALCULAR LA CAPACIDAD REAL DEL CAMIÓN
* Tenemos que tener en cuenta que el material tiene una densidad de 1800 kg/m3 (1.8 Tm/m3), por lo que el camión puede cargar como máximo 229 toneladas métricas y no sobrepasar su capacidad.
* Entonces tendremos que dividir máxima de la tolva entre la capacidad del cucharón del cargador:

```python
capacidad_carga = float(input("Digite la capacidad de carga en Tm: "))
densidad = float(input("Digitela densidad: "))
n = capacidad_carga / (q * densidad)
N = round(n)
N
```

* Se admitirá 5 ciclos de carga, ya que si se realiza una más, estaremos sobrecargando al camión.
* La producción por cilco del camión será:

```python
Q = round(n) * q
Q
```

* Si vemos desde el punto de vista del volumen, el camión estará al 68.18% de su capacidad nominal de 176 m3.

```python
QR = Q * 1.8
print(f"La capacidad real es de {QR} toneladas metricas")
```

## FEEDBACK: USO DE LA GRÁFICA DE RENDIMIENTO

## PASO 8: REALIZAR EL CÁLCULO DEL TIEMPO DE RETORNO
* El retorno del botadero a la obra lo realiza sin carga (vacia), por lo que el peso del camión es de 154.73 toneladas, entonces se realiza la misma gráfica tomando el lado vacio.

## PASO 9: CALCULAR EL TIEMPO TOTAL Y DE PRODUCCIÓN
* Se tiene el:
    * Tiempo de carga: 3 minutos 
    * Tiempo de acarreo: 8.79 minutos
    * Tiempo de descarga: 1 minuto
    * Tiempo de retorno: 2.84 min
    * La duración del ciclo es Cm = 15.6 minutos
* Para las condiciones normales de mantenimiento y de operación, la eficiencia es de 65%
* La producción horaria y por turno será:

```python
Cm = float(input("Digite la duración del ciclo en minutos: "))
efi = float(input("Digite la eficiencia: "))
d = float(input("Digite la densidad: "))
tiempo_turno = float(input("Digite el tiempo por turno: "))
produccion_h = Q * 60 / Cm * efi
produccion_h_t = produccion_h * d
produccion_h_tm = produccion_h_t * tiempo_turno
print(f"{produccion_h} m3/h \n{produccion_h_t} Tm/h \n{produccion_h_tm} Tm/turno")
```
