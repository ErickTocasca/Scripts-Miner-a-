## Ciclo de excavación (pala)
* Fase de Excavación: Incluye empujar el balde en el banco, levantar el balde para llenarlo y retraer  el balde del banco. Balance adecuado de fuerzas de empuje y levante.
* Fase de Giro: Comienza cuando el balde libra el banco vertical y horizontalmente. Se controla la posición del balde mediante una trayectoria de giro y altura de descarga planeada.
* Fase de Descarga: Comienza antes de que el balde cargado pase por el lado del camión de carga y termina cuando el movimiento de giro para y retrocede para regresar al banco. 
* Fase de Retorno: Incluye el giro del bastidor superior en dirección al banco y también baja el balde en posición retraída, con el fin de cerrar la tapa.

## Empuje y Levante
Son fuerzas que influyen directamente al rendimiento del carguio de una pala.
* Crowd (empuje)
* Retract (Retrair)
* Hoist (Izaje)
### Levantar (Boom Jack)
* El boom jack es el resultado de la excesiva fuerza de empuje aplicada durante la fase de levante el ciclo de excavación.
* Ocurre cuando se aplica una fuerza de empuje excesiva sin aplicar simultáneamente una fuerza proporcional de levante.

## Giro y Tiempo de Ciclo de Pala
### Giro
* La fase de giro comienza cuando el balde libra el banco vertical y horizontal.
* Acelere suavemente a través de la primera mitad del arco de giro, luego comience a desacelerar.
### Tiempo de ciclo de Pala 
**DEPENDE DE CIERTOS FACTORES:**
* Tipo de escificaciones de pala.
* Velocidad y aceleración de giro y de levantamiento.
* Ubicación de camiones en relación a la pala (ángulos de giro).
* Tiempos de espera, acomodación y atraso.
* Habilidad del operador.

## Características del balde
### El balde / Cucharón / Lampón
Incrementar el ángulo de ataque pone mas trabajo en la escavación, resultando en mejorar la eficiencia y reducir el consumo de energía.

### Capacidad  del Balde
* La capacidad del balde se determina por el peso de trabajo del balde más el peso promedio del material en el balde. El peso del balde más el peso del material es igual a la carga suspendida de la pala.
* Siempre y cuando la carga suspendida no sea mayor que la carga nominal suspendida, la pala puede trabajar.

**CUANDO SE CARGA LA PALA A UN NIVEL MAYOR A LA CARGA NOMINAL SUSPENDIDA, SURGEN VARIOS PROBLEMAS TALES COMO**
* Fisuras en la pluma.
* Fatiga del mango.
* Fisuras de la tornamesa.
* Problemas con el tren de polines.
* Fallas prematuras en las zapatas de oruga.
* Vida útil reducida de los engranajes.
* Sobrecalentamiento de motores.
* Desgaste de frenos de izaje y giro.
### Capacidad del Balde
* Los baldes se llenan generalmente entre 90% y 95% de la capacidad rasa SAE. En la actualidad estamos viendo que los baldes normalmente se llenan entre el 105% y 110% de su capacidad rasa SAE, salvo algunas excepciones. Va depender del material, del frente, de la habilidad del operador.

### Desventajas del Double Back Up
* Se requiere grandes habilidades de operadores de camiones.
* La mitad de los camiones son cargados por el lado ciego.
* Es necesaria un área adecuada de minado.
* No aplicable para frentes de trabajo bajos que requieren movimientos frecuentes de pala.
* Comparado con Drive-by:
    * Los movimientos requieren más asistencia de tractores.
    * El ángulo de pit es más dificil de mantener.

## Single Back Up ( Cargar a un solo lado)
* Para áreas de minado restringida.
* La forma irregular de las caras es fácil de acomodar.
* Los bolsillos de material pueden ser minados selectivamente.
* Cargado solamente por el lado del operador.
* Para el camión es fácil ubicarse usando la pala como guía.

## Drive By (Avance Paralelo al Banco)
* Maniobras mínimas para la unidad de carga.
* Los movimientos de pala son mínimos a través del corte.
* La superficie del borde puede ser usada como lecho de vía.
* El ángulo del piso del pit es fácil de mantener.
* Los camiones no requieren retroceder, cusando así menos daños de neumáticos.

### Carga Doble
* Espacio de trabajo reducido.
* Máquina muy productiva.
* Configuración y excavación en ángulo recto.
### Carga Simple
* Menos limpieza del suelo.
* Más facil de mantener.
* Configuración y excavación en ángulo recto.
### Elementos que debe tener en cuenta
* Posiciones previas del camión cargador.
* Posciones del camión cargador en el radio de giro de la pala.
* El camión debe estar ubicado sobre suelo limpio.
* Menos tiempo de carga de camiones.
* Limpie el suelo después de cada movimiento de la pala.
