# CASO 2: CALCULO DE PRODUCCIÓN DE UN CARGADOR FRONTAL
**Calcular la producción horario de un cargador sobre ruedas CAT 980K, si este tuviera las siguientes condiciones:**
* Material suelo arenoso.
* Densidad del material 1900 kg/m3.
* Tiempo de ciclo de carga en V: 0.6 min.   (Pala ciclo normal de menor tiempo)
* Condicones normales de operación y mantenimiento: 0.65  (Eficiencia)

## PASO 1: SELECCIONAR EL CUCHARÓN:
**ESTO OBTENERMOS DEL CUADRO DE FABRICANTE**
* Para maximizar la producción del cargador frontal 980K (marca Caterpillar de carga estándar con un material de 1900 kg/m3 de densidad), se elige un cucharón de 4.4 m3.

```python
from IPython.display import Image
Image(filename="../Catalago_sel_cucharon.jpg")
```

## PASO 2:
* La capacidad colmada con este cucharón para este material tendrá que multiplicarse por 110%, siendo la fórmula general:

```python
q_1 = 1.1 * 4.4   # 110%
print(f"{round(q_1,2)} m3")
```

## PASO 3: CALCULAR EL FACTOR DE CARGA (k) Y PRODUCCIÓN POR CICLO (q)

```python
Image(filename="../condiciones_de_carga.jpg")
```

* El material es suelo arenoso, porque la condición de carga es promedio.
* El factor de carga: k = 0.8
* La producción por ciclo es: **q = q_1 x k**
* Donde:
    * q1: Capacidad colmada (m3).
    * k: Factor de carga (tabla).

```python
q = q_1 * 0.8
print(f"{round(q,2)} m3")
```

## PASO 4: CALCULAR LA PRODUCCIÓN HORARIA (Q)
* Para un equipo en condiciones normales de operación y mantenimiento, la eficiencia será 65% (a tenido reparaciones frecuentes, poca utilización o es un equipo auxiliar o de apoyo).
* La producción horaria del cargador se calcula mediante: **Q = q x 60 x E / Cm**
    * Q: Producción horaria (m3/h)
    * q: Capacidad del lampón (m3)
    * E: Eficiencia
    * Cm: Tiempo de ciclo (min)

```python
Q = q * 60 * 0.65 / 0.6
print(f"{round(Q,2)} m3/h")
```
